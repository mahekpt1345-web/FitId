```python
from FitTrack.WorkoutTracker import WorkoutTracker
from FitTrack.NutritionTracker import NutritionTracker
from FitTrack.HealthMetrics import HealthMetrics
from FitTrack.FitnessGoals import FitnessGoals
from FitTrack.DataVisualization import DataVisualization

# Initialize all trackers
workout_tracker = WorkoutTracker()
nutrition_tracker = NutritionTracker()
health_metrics = HealthMetrics()
fitness_goals = FitnessGoals()
data_visualization = DataVisualization(workout_tracker, nutrition_tracker, health_metrics, fitness_goals)

# --- Workout Tracker Tests ---
print("\n--- Testing WorkoutTracker ---")
workout_tracker.add_workout(name="Running", duration=30, calories_burned=300, date="2025-04-05")
print("Workouts:", workout_tracker.view_workouts())

# --- Nutrition Tracker Tests ---
print("\n--- Testing NutritionTracker ---")
nutrition_tracker.add_meal(meal_name="Oatmeal", calories=250, macros={"proteins": 8, "fats": 5, "carbs": 40})
print("Daily Meals:", nutrition_tracker.view_daily_meals())
print("Total Calories:", nutrition_tracker.calculate_daily_calories())

# --- Health Metrics Tests ---
print("\n--- Testing HealthMetrics ---")
health_metrics.log_weight(weight=70, date="2025-04-05")
print("Recent Weight:", health_metrics.get_recent_weight())
bmi = health_metrics.calculate_bmi(weight=70, height=1.75)
print("BMI:", bmi)

# --- Fitness Goals Tests ---
print("\n--- Testing FitnessGoals ---")
fitness_goals.set_goal(goal_type="weight_loss", target=5)
fitness_goals.update_goal_progress(goal_type="weight_loss", progress=2)
print("Fitness Goals:", fitness_goals.view_all_goals())

# --- Data Visualization (Mock Execution) ---
print("\n--- Testing DataVisualization ---")
data_visualization.generate_workout_chart()  # Should display workout trend
data_visualization.generate_macronutrient_pie()  # Should visualize macronutrient distribution

# --- Edge Case Tests ---
print("\n--- Testing Edge Cases ---")
print("Invalid BMI (Height = 0):", health_metrics.calculate_bmi(weight=70, height=0))
print("Nonexistent Goal Remaining:", fitness_goals.get_remaining_goal(goal_type="nonexistent_goal"))
```
